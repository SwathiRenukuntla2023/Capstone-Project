```python
import pandas as pd

# load the data from the file to dataframe

Movie_Data = pd.read_csv("DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv")
Movie_Data
```

```python
# Check the available column names in the DataFrame

print(Movie_Data.columns)
```

```python
# Extracting the Json values

import json

# Extract genres from the "genres" column

genres = []
for row in Movie_Data['genres']:
    # Check if the row is not NaN
    if pd.notnull(row):
        # Parse the JSON data
        data = json.loads(row)
        
        # Extract the genre names
        genre_names = [genre['name'] for genre in data]
        
        # Append the genre names to the genres list
        genres.append(genre_names)
    else:
        # Append None if the row is NaN
        genres.append(None)

# Replace the "genres" column with the extracted genre values
Movie_Data['genres'] = genres
```

### Task - 10 :

```python
# Sort the data in descending order based on revenue& filter the top_500

Top_500_Movies = Movie_Data.sort_values("revenue", ascending = False).head(500)
Top_500_Movies
```

```python
import seaborn as sns

# Columns to analyze

columns_to_analyze = ['budget', 'revenue', 'runtime']

# Calculate measures of central tendency

central_tendency = {}

for col in columns_to_analyze:
    Mean = Top_500_Movies[col].mean()
    Median = Top_500_Movies[col].median()
    Mode = Top_500_Movies[col].mode()[0]
    
    central_tendency[col] = {'Mean': Mean, 'Median': Median, 'Mode': Mode}

# Display measures of central tendency

print("Measures of Central Tendency:")
print(pd.DataFrame(central_tendency))

# Create subplots for outlier analysis

fig, axes = plt.subplots(1, len(columns_to_analyze), figsize=(16, 9))

for i, col in enumerate(columns_to_analyze):
    sns.boxplot(x = col, data = Top_500_Movies, ax = axes[i])
    axes[i].set_title(f"Boxplot for Outlier Analysis - {col}")

plt.tight_layout()
plt.show()
```
